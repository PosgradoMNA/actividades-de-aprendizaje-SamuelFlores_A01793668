**Exportación e importación de datos en python**

```python
import numpy as np      #Importamos las librerias
import pandas as pd
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"  #Definimos la ruta del csv
```

```python
df = pd.read_csv(url, header = None)   #Leemos y alamacenamos el contenido del csv en df
df.head(5)   #Mostramos las primeras lineas del dataframe
```

```python
df.tail(5) #Mostramos las ultimas lineas del dataframe
```

```python
#Definimos el nombre de cada columna y lo almacenamos en headers

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
df.columns = headers       #Le damos el nombre de las columnas contenidas en header
df.head(10)
```

```python
df1=df.replace('?',np.NaN)  #Reemplazamos el simbolo de interrogacion por Nan para que pueda ser reconocido como dato vacio
df=df1.dropna(subset=["price"], axis=0)  #Eliminamos los datos faltantes de la columna precio
df.head(20)
```

```python
print("The last 10 rows of the dataframe\n")
df.tail(10)
```

```python
df.to_csv("automobile.csv", index=False) #Guardamos el dataframe como un csv
```
