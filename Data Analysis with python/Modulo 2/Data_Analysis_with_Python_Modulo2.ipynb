# **Preprocesamiento de datos**

El preprocesamiento de datos es un paso necesario para el análisis de los datos, y este mismo se define como el proceso de convertir los datos de un formato inicial hacia otro formato ordenado y preparado para su análisis. (Limpieza de datos).

**Objetivos:**

Identificar y manipular valores faltantes.

Formateo de datos.

Normalización de datos (escalado)

Copnvertir valores categóricos a numéricos.



Los datos faltantes se pueden representar como:

1-?

2-N/A

3-O 

4-Espacio en blanco

# **Lidiando con valores faltantes en los registros de datos:**

-Investigar los valores faltantes para completar los datos.

-Eliminando valores faltantes:

Eliminar por completo la variable (Si son muchos registros a los cuales les falta una variable en específico).

Eliminar los registros con valores faltantes. (si no son tantos valores faltantes en una misma variable)

-Remplazar los datos faltantes:

Remplazar por promedio de la columna.

Remplazar por la moda de la columna.

Remplazar por medio de alguna función

-Dejarlo como dato faltante.

```python
import pandas as pd                    #Importamos librerias
import matplotlib.pylab as plt
import numpy as np
```

```python
#Almacenamos los datos en filename
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv" 
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",             #Le damos nombre a las columnas
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)   #Creamos el dataframe
```

```python
df.head(5)   #Mostramos los datos del dataframe df
```

```python
df.replace("?", np.nan, inplace = True)  #Remplazamos los simbolos por nan
```

```python
missing_data = df.isnull()     #Mostramos los valores vacios (observamos 5 registros con datos vacios)
missing_data.head()
```

```python
for column in missing_data.columns.values.tolist():         #Contamos los datos vacios de cada columna del dataframe
    print(column)
    print (missing_data[column].value_counts())
    print("")  
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)     #Calculamos el promedio de la columna horsepower para sustituir los valores vacios
print("Average horsepower:", avg_horsepower)
```

```python
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)   #Remplazamos los datosa vacios por el promedio
```

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)           #Calculamos el promedio de la columna peak rmp para sustituir los valores vacios
print("Average peak rpm:", avg_peakrpm)
```

```python
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)     #Remplazamos los datosa vacios por el promedio
```

```python
df.dropna(subset=["price"], axis=0, inplace=True)  #Eliminamos todos los registros con datos vacios en precio
```

```python
df.head()  #Mostramos el dataframe
```

```python
df.dtypes   #Mostramos el tipo de dato de cada columna
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")      #Convertimos los datos para darle un formato adecuado
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")                               
```

```python
df.dtypes   #Tipos de datos despues de conversion
```

```python

df['city-L/100km'] = 235/df["city-mpg"]   #Convertimos los datos de millas por galon a litros por 100 km

# check your transformed data 
df.head()
```

# **Question 1**
Based on the example above, replace NaN in "stroke" column with the mean value.

```python
#Calculamos promedio
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average of stroke:", avg_stroke)

# remplazamos nan por promedio
df["stroke"].replace(np.nan, avg_stroke, inplace = True)
```

# **Question 2:**
According to the example above, transform mpg to L/100km in the column of "highway-mpg" and change the name of column to "highway-L/100km".

```python
#Convertimos los datos de millas por galon a litros por 100 km
df["highway-mpg"] = 235/df["highway-mpg"]

# renombramos la columna
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)

# Mostramos el resultado
df.head()
```

```python
df['length'] = df['length']/df['length'].max()       #Normalizamos las variables escalandolas
df['width'] = df['width']/df['width'].max()
```

# **Question 3:**
According to the example above, normalize the column "height".

```python
df['height'] = df['height']/df['height'].max()   #Normalizamos la columna altura

# Mostramos las columnas escaladas
df[["length","width","height"]].head()
```
