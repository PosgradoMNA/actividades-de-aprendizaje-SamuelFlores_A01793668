```python
import pandas as pd   #Importamos librerias
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Definimos funcion de grafica de distribucion
def DistributionPlot(RedFunction, BlueFunction, RedName, BlueName, Title):
    width = 12
    height = 10
    plt.figure(figsize=(width, height))
    
    #Definimos parametros 
    ax1 = sns.distplot(RedFunction, hist=False, color="r", label=RedName)
    ax2 = sns.distplot(BlueFunction, hist=False, color="b", label=BlueName, ax=ax1)
    
    #Graficamos
    plt.title(Title)
    plt.xlabel('Price (in dollars)')
    plt.ylabel('Proportion of Cars')

    plt.show()
    plt.close()
```

```python
#Definimos funcion de grafica polinomial
def PollyPlot(xtrain, xtest, y_train, y_test, lr,poly_transform):
    width = 12
    height = 10
    plt.figure(figsize=(width, height))
    
    
    #training data 
    #testing data 
    # lr:  linear regression object 
    #poly_transform:  polynomial transformation object 
 
    xmax=max([xtrain.values.max(), xtest.values.max()])

    xmin=min([xtrain.values.min(), xtest.values.min()])

    x=np.arange(xmin, xmax, 0.1)

    
    #Graficamos
    plt.plot(xtrain, y_train, 'ro', label='Training Data')
    plt.plot(xtest, y_test, 'go', label='Test Data')
    plt.plot(x, lr.predict(poly_transform.fit_transform(x.reshape(-1, 1))), label='Predicted Function')
    plt.ylim([-10000, 60000])
    plt.ylabel('Price')
    plt.legend()
```

#Ridge Regression

```python
pr=PolynomialFeatures(degree=2) #Aplicamos transformacion polinomial de grado 2
x_train_pr=pr.fit_transform(x_train[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg','normalized-losses','symboling']]) #Entrenamiento
x_test_pr=pr.fit_transform(x_test[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg','normalized-losses','symboling']]) #Prueba
```

```python
from sklearn.linear_model import Ridge #Importamos libreria
```

```python
RigeModel=Ridge(alpha=1) #Creamos objeto y le asignamos parametro de regularizacion
```

```python
RigeModel.fit(x_train_pr, y_train) #Aplicamos fit
```

```python
yhat = RigeModel.predict(x_test_pr) #Predecimos salida
```

```python
print('predicted:', yhat[0:4]) #Comparamos las primeras 4 predicciones
print('test set :', y_test[0:4].values)
```

```python
#Escogemos el valor de alpha que minimize el error de prueba

from tqdm import tqdm #Importamos libreria

Rsqu_test = []
Rsqu_train = []
dummy1 = []
Alpha = 10 * np.array(range(0,1000))
pbar = tqdm(Alpha)

for alpha in pbar:
    RigeModel = Ridge(alpha=alpha)  #Definimos modelo
    RigeModel.fit(x_train_pr, y_train) #Entrenamos
    test_score, train_score = RigeModel.score(x_test_pr, y_test), RigeModel.score(x_train_pr, y_train)
    
    pbar.set_postfix({"Test Score": test_score, "Train Score": train_score})

    Rsqu_test.append(test_score)
    Rsqu_train.append(train_score)
```

```python
#Graficamos 

width = 12
height = 10
plt.figure(figsize=(width, height))

plt.plot(Alpha,Rsqu_test, label='validation data  ')
plt.plot(Alpha,Rsqu_train, 'r', label='training Data ')
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.legend()
```
